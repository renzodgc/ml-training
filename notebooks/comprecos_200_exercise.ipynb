### Without filters in place, what is the mean of means confidence of normal vs '200' version of recommendations for top K (10) recommendations?


```python
normal_confidences_top_10_without_lift_filter = df_without_lift_filter.sort_values(
    "confidence", ascending=False
).groupby(
    ["pc9_prod"]
).head(
    10
).groupby(
    ["pc9_prod"]
)["confidence"].mean().mean()
print(normal_confidences_top_10_without_lift_filter)

confidences_top_10_populare_recommendations_unfiltered = unfiltered_products_that_recommend_a_popular_product.sort_values(
    "confidence", ascending=False
).groupby(
    ["pc9_prod"]
).head(
    10
).groupby(
    ["pc9_prod"]
)["confidence"].mean().mean()
print(confidences_top_10_populare_recommendations_unfiltered)
```
